```python
# Loops

# Loops in Python are used to repeat a block of code multiple times without
# writing the same code again and again.

# Loops help us repeat tasks efficiently and smartly.

# A loop is a control structure that repeats a block of code until a condition is met.

 # for and while

# for loop
# it is used to iterate over a Sequence [string , list , tuple , array etc...]

# for iterable_variable in Sequence:
#   code to be executed

# AFTER EVERY ITERATION BODY OF FOR LOOP WILL BE EXECUTED

for i in "Python":
  print(i)
```

```python
# range() function
# it is used to generate a sequence of integer values

# range(x)
# it is used to generate a sequence of integer values
# starting from Zero till x-1

print(list(range(11)))
print(list(range(30)))

# range(x,y)
# it is used to generate a sequence of integer values
# starting from x till y-1

print(list(range(1,11)))
print(list(range(10,70)))


#range (x,y,stepCount)
# x till y-1 based on stepCount

print(list(range(1,11,4)))
```

```python
# Nested for Loop
# for loop defined inside another for loop

colours = ["red", "blue", "green"]
vehicles = ["car", "bus", "bike"]

# outer loop will wait untill inner loop finishes iteration

for c in colours:
  for v in vehicles:
    print(c,v)
```

```python
rows = 5
for i in range(1,rows+1):
  for j in range(i):
    print('*',end=" ")
  print()
```

```python
# while loop
# while loop repeats a block of code
# until a certain condition met

# syntax

# initialization
# while condition:
#  a block of code
#  increment / decrement

# While checks the condition first
# if the condition is TRUE while will execute the block of code
# after executing the block of code while will check the condition again


i = 1
while i<=3:    # 1<=3 (t)           2<=3(t)             3<=3(t)         4<=3(F)
  print(i)     # 1                  2                   3
  i += 1       # i = 1 + 1 (2)      i = 2 + 1 (3)       i = 3 + 1 (4)
```

```python
# print mathemetical table of a number using while loop

num = int(input('Enter Any Number : '))
i = 1
while i<=10:
  print(f"{num} x {i} = {num*i}")
  i += 1
```

```python
#use while loop when the number of iterations are not known

num = 121
sum = 0

while num!=0:               # 1st iterations 121!=0(t)      # 2nd iterations 121!=0(t)    # 3rd iterations 121!=0(t)
  rem = num % 10            #rem = 1                         #rem = 2                     #rem = 1
  sum = sum*10 + rem        #sum = 1                        #sum = 12                     #sum = 121
  num = num // 10           #num = 12                       #num = 1                      #num = 0

if num == sum:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# break and continue statement

# break statement
# it is used to terminate the execution of a LOOP

# break and continue should be defined inside decision making

for i in range (1,11):
  if i == 5:
    break
  print(i)
```

```python
# continue statement
# it is used to skip the current execution of a LOOP

for i in range (1,11):
  if i == 5 or i == 9 :
    continue
  print(i)
```
