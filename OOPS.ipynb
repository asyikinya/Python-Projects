```python
# OOPS
# OOP stands for Object-Oriented Programming.
# Object-oriented programming (OOP) is a programming paradigm based on the
# concept of "objects"
 
# Procedural programming is about writing procedures or methods that perform operations
# on the data, while object-oriented programming is about creating objects
# that contain both data and methods.
 
# Object-oriented programming has several advantages over procedural programming:
 
# a.) OOP is faster and easier to execute
# b.) OOP provides a clear structure for the programs
# c.) OOP helps to keep the Java/Python code DRY "Don't Repeat Yourself", and makes the code easier
#     to maintain, modify and debug
# d.) OOP makes it possible to create full reusable applications with less code and
#     shorter development time
 
# Class
# A class is a blueprint that defines the variables and the methods common to all objects.
# a class is a collection of methods and variables
 
 
# Object
# An object is called an instance of a class.
 
 
# Self Keyword
# The self parameter is a reference to the current instance of the class, and is used to
# access variables that belongs to the class.
# The first parameter of each method in a class must be the self ,
# which refers to the calling object. However, you can give any name to the first
# parameter, not necessarily self.
 
# __init__ in Python
#  Constructors are used to initializing the object’s state. The task of constructors is
# to initialize(assign values) to the data members of the class when
# an object of the class is created.
# The __init__() function is called automatically every time the class is being used to create
# a new object.
```

```python
# class in Python
# a class is a collection of methods/functions and variables
# in python a class can be defined by using a class keyword

class myClass:

    '''
    This is my first class in python.
    I have just started learning OOPs
    '''

    a = 10     # member variable
    b = 20

    def func1():   # member function
        print('I am the method of a class')

    def func2():
        print('I am the another method of a class')
```

```python
class myClass:

    a = 10  
    b = 20

    def func1():
        print('I am the method of a class')

    def func2():
        print('I am the another method of a class')

# In OOPs in order to access class variables and functions 
# we need to define the Object of a class

# syntax for defining the object
# objName = className()

# object can access the class variables and functions
# in python a class can have n number of objects

obj1 = myClass()  # creating an object of a class
obj2 = myClass()  # creating an another object of a class

print(obj1.a)
print(obj1.b)

print(obj2.a)
print(obj2.b)
```

```python
class myClass:

    a = 10  
    b = 20

    def func1():
        print('I am the method of a class')

    def func2():
        print('I am the another method of a class')

obj1 = myClass()  
obj2 = myClass() 

# A CLASS SHARE COMMON PROPERTIES AMONG ALL OBJECTS

print(obj1.a , obj2.a)
print(obj1.b , obj2.b)
```

```python
# Self Keyword
# IN PYTHON WHEN WE DEFINE A FUNCTION INSIDE A CLASS
# THE FIRST PARAMETER SHOULD BE SELF
 
# function without parameter inside a class
# def func1(self):
#   pass
 
# function with parameter inside a class
# def func1(self,a,b):
#   pass
 
# SELF REPRESENT THE CURRENT CALLING OBJECT
# OR YOU CAN SAY SELF KEYWORD HELP A CLASS TO IDENTIFY WHICH OBJECT IS
# CURRENTLY ACCESSING THE FUNCTION
```

```python
class myClass:   # 3.) giving your class a clear idea that now obj1 wants to use this function
 
    a = 10  
    b = 20
 
    def func1(self): # 2.) self value will be replaced by the current calling object  func1(obj1)
        print('I am the method of a class')
 
    def func2(self):
        print('I am the another method of a class')
 
obj1 = myClass()  
obj2 = myClass() 
obj3 = myClass()
 
obj1.func1()  # 1.) here i am calling this func1 with obj1 
obj2.func1()
obj3.func1()
```

```python
# Task
# We need to take 3 information from 3 users
# user1          user2            user3
# name = ?      name = ?         name = ?  
# email = ?     email = ?       email = ?
# mobno = ?     mobno = ?       mobno = ?

# Task 1 : take 3 information from 3 users
# Task 2 : print all these information
# Task 3 : print only the information of second user  --> user2 name email and mobno
# Task 4 : print the email and mobno of 3rd user

# Solution 1 : Using Procedural Programming

users = []

for i in range(1,4):   # 1 2 3
    print(f"Enter Details For User {i}")    # Enter user1         Enter user2 
    name = input('Enter Your Name : ')      # john                alex
    email = input('Enter Your Email : ')    # j@gmail.com         al@gmail.com
    mobno = input('Enter Your MobNo : ')    # 60327266            607895544

    # store this information
    user_data = [name , email , mobno]    # user_data = [alex al@gmail.com 607895544]
    users.append(user_data)               # users [ [john j@gmail.com 60327266] ,
                                          #         [alex , al@gmail.com , 607895544]           
                                          #      ]
    
# Task 2 : print all these information
print(users)

# Task 3 : print only the information of second user  --> user2 name email and mobno
print(users[1])

# Task 4 : print the email and mobno of 3rd user
print(users[2][1])
print(users[2][2])     # 12 LOC
```

```python

class UserDetails:

    def __init__(self, name, email, mobno):
        self.name = name
        self.email = email
        self.mobno = mobno

    def printUser(self):
        print("Name  :", self.name)
        print("Email :", self.email)
        print("MobNo :", self.mobno)
        print("----------------------")

user1 = UserDetails("NURUL", "nurul@gmail.com", "0135947814")
user2 = UserDetails("ahmad", "ahmad@gmail.com", "0123456789")
user3 = UserDetails("zainal", "zainal@tremcocpg.com", "01236547899")

print("===== USER DETAILS =====")

user1.printUser()
user2.printUser()
user3.printUser()
```

```python
#use in it & enumerate

class UserDetails:

    def __init__(self, name, email, mobno):
        self.name = name
        self.email = email
        self.mobno = mobno

    def printUser(self):
        print("Name  :", self.name)
        print("Email :", self.email)
        print("MobNo :", self.mobno)


users = [
    UserDetails("NURUL", "nurul@gmail.com", "0135947814"),
    UserDetails("ahmad", "ahmad@gmail.com", "0123456789"),
    UserDetails("zainal", "zainal@tremcocpg.com", "01236547899")
]

print("===== USER DETAILS =====\n")

for i, user in enumerate(users, start=1):
    print(f"User {i}")
    user.printUser()
    print("----------------------")
```
