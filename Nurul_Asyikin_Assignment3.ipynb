```python
# Q1. Python program to count positive and negative numbers in a list.
# a = [10, -20, 30, -40, 50, -60, 0]

#SOLUTION

a = [10, -20, 30, -40, 50, -60, 0]

positive = 0
negative = 0

for num in a:
    if num > 0:
        positive += 1
    elif num < 0:
        negative += 1

print("Positive numbers:", positive)
print("Negative numbers:", negative)
```

```python
# Q2. Using List Comprehension. Write a Python program to count positive and negative numbers in a list.
# a = [10, -20, 30, -40, 50, -60, 0]

#SOLUTION

a = [10, -20, 30, -40, 50, -60, 0]

positive_count = len([x for x in a if x > 0])
negative_count = len([x for x in a if x < 0])

print("Positive numbers:", positive_count)
print("Negative numbers:", negative_count)
```

```python
# Q3. Remove Multiple Elements from List in Python
# a = [10, 20, 30, 40, 50, 60, 70]
# remove = [20, 40, 60]
# Resulting list = [10, 30, 50, 70]

#SOLUTION

a = [10, 20, 30, 40, 50, 60, 70]
remove = [20, 40, 60]

result = []

for x in a:
    if x not in remove:
        result.append(x)

print("Resulting list =", result)
```

```python
# Q4. Remove duplicates from list in Python
# a = [1, 2, 2, 3, 4, 4, 5]

#SOLUTION
#USING SET() - set() automatically removes duplicates.

a = [1, 2, 2, 3, 4, 4, 5]

result = list(set(a))

print(result)
```

```python
# Q5. Python program to find the sum of all items in a dictionary
# Input: {'a': 100, 'b': 200, 'c': 300}
# Output: 600

#SOLUTION - Using sum()

data = {'a': 100, 'b': 200, 'c': 300}

total = sum(data.values()) # → gives all the values

print("Sum of all items:", total)#sum() → adds them together
```

```python
# Q6. Key with Maximum Unique Values - Python
# dict1 = {"key1": [5, 7, 5, 4, 5], "key2": [6, 7, 4, 3, 3], "key3": [9, 9, 6, 5, 5]}
# output : key2

#SOLUTION

dict1 = {
    "key1": [5, 7, 5, 4, 5],  #Removes duplicates
    "key2": [6, 7, 4, 3, 3],
    "key3": [9, 9, 6, 5, 5]
}

max_key = ""
max_count = 0

for key, value in dict1.items():
    unique_count = len(set(value))  #Counts unique values

    if unique_count > max_count:    #Compare counts
        max_count = unique_count
        max_key = key

print("Key with maximum unique values:", max_key)
```

```python
# Q7. Story: An online store records daily product sales.
# Convert this list into total sales per product.
sales = [
    ("apple", 3),
    ("banana", 2),
    ("apple", 5),
    ("banana", 1),
    ("orange", 4)
]

# Output : {'apple': 8, 'banana': 3, 'orange': 4}

#SOLUTION

#GIVEN DATA
sales = [
    ("apple", 3),
    ("banana", 2),
    ("apple", 5),
    ("banana", 1),
    ("orange", 4)
]

total_sales = {}

for product, quantity in sales:
    if product in total_sales:
        total_sales[product] += quantity
    else:
        total_sales[product] = quantity

print(total_sales)
```

```python
# Q8. Undo Feature in Text Editor
# A text editor supports an undo operation.
# Given operations:
# actions = ["type A", "type B", "delete", "type C"]
# Output : ['type A', 'type C']

# actions = ["type A", "type B", "delete", "type C","type B","delete","type A","type C","delete"]
# Output : ['type A', 'type C', 'type A']

#SOLUTION

actions = ["type A", "type B", "delete", "type C","type B","delete","type A","type C","delete"]

result = []

for action in actions:
    if action == "delete":
        if result:
            result.pop()
    else:
        result.append(action)

print("Output", result)
```

```python
# Q9. Print the index number of those Two boxes whose weights sum to a target.
# weights = [2, 7, 11, 15 , 6 , 3, 7 , 3 , 7 , 2]
# target = 9

# Output
'''
0 1
4 5
0 6
4 7
0 8
8 9
'''

#SOLUTION
#Check every pair of weights.
#If sum equals target → print their index numbers.

weights = [2, 7, 11, 15 , 6 , 3, 7 , 3 , 7 , 2]
target = 9
seen = {}

for i, w in enumerate(weights):
    if target - w in seen:
        print(seen[target - w], i)
    seen[w] = i
```

```python
# Q10. Online Store Price Update (map + lambda)
# An online store increases all product prices by 10%.
# prices = [100, 200, 300, 400]
# output [110, 220, 330, 440]

#SOLUTION

prices = [100, 200, 300, 400]

updated_prices = list(map(lambda x: x * 1.10, prices)) #map() applies function to every item. lambda x: x * 1.10 → increase by 10%.

print(updated_prices)
```

```python
 #Q11. Employee Bonus System (map + if–else)
# Employees earning ≥ 50,000 get a 20% bonus, others get 10%.
# salaries = [40000, 60000, 50000, 30000]

#SOLUTION
salaries = [40000, 60000, 50000, 30000]

bonus = list(map(lambda x: x * 1.20 if x >= 50000 else x * 1.10, salaries)) # if salary ≥ 50000 → add 20%, if less → add 10%

print(bonus)
```

```python
# Q12. Order Filtering System (filter + multiple conditions)
# An e-commerce system wants orders above ₹1000 and not cancelled.

orders = [
    {"amount": 1200, "status": "completed"},
    {"amount": 800, "status": "completed"},
    {"amount": 1500, "status": "cancelled"},
    {"amount": 2000, "status": "completed"}
]

#SOLUTION

#amount > 1000 , status is "completed"

orders = [
    {"amount": 1200, "status": "completed"},
    {"amount": 800, "status": "completed"},
    {"amount": 1500, "status": "cancelled"},
    {"amount": 2000, "status": "completed"}
]

filtered_orders = list(filter(
    lambda x: x["amount"] > 1000 and x["status"] == "completed",
    orders
))

print(filtered_orders)
```

```python
# Q13. Data Cleaning Pipeline (map + filter + lambda)
# From a data pipeline:
#  a.)Remove negative numbers
#  b.) Square remaining numbers
#  c.) Keep only results > 50

# data = [-4, -1, 2, 4, 8]
# Output : [64]

#SOLUTION

#a) Remove negative numbers, b) Square numbers, c) Keep results > 50

data = [-4, -1, 2, 4, 8]

result = list(
    filter(lambda x: x > 50, map(lambda x: x**2,filter(lambda x: x >= 0, data)) )
)

print(result)
```
