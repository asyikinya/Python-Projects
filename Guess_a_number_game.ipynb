```python
# Guess a number game
# your machine will generate a random number
# we need to guess that number in 4 attempts

import random
guessAttempt = 1
print('Guess A Number Game 🤖')
print('Enter Your Name')
name = input()
print(f'Welcome User {name} 😎')
print('I am generating one random number between 1 till 13 👻 Please Wait.....')
num = random.randint(1,13)
print('✅ Number Is Generated. Guess The Number in 4 Attempts')

while guessAttempt <= 4:
  print(f'\n{guessAttempt} Attempt. Enter Your Guess.')
  user_guess = int(input())

  if user_guess == num:
    print(f'Congratulation 🤩 You Guess The Number Right In {guessAttempt} Attempt.')
    break

  if user_guess > num:
    print('⬆️ Your Guess Is Too High. Think Of A Lower Number\n')

  if user_guess < num:
    print('⬇️ Your Guess Is Too Low. Think Of A Higher Number\n')

  guessAttempt = guessAttempt + 1

if user_guess != num:
  print('🛑 Zero Attempt Left. Game Over ! ')
  print(f'The Correct Number Was 📢 {num}')
```
