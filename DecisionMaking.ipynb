```python
# if else

# if used to check the condition first
# if the condition is TRUE it will execute the block of code

# if the condition is FALSE , then else part will be executed

age=int(input("Enter Your Age : "))
if age>=18:
  print("Welcome User")
  print("Eligible To Play The Game")
  print("Game is Loading")
else:
  print("Sorry ~!")
  print("Not Eligible To Play Game")
```

```python
# if elif else

# ONLY ONE BLOCK OF CODE WILL BE EXECUTED AT A TIME

num = int(input('Enter Any Number : '))
if num>0:
  print('The Number Is Positive')
elif num == 0:
  print('The Number Is Zero')
else:
  print('The Number Is Negative')
```
